###### Popsize

```python
# Задаем список значений параметра, а также переменные для хранения результатов выполнения оптимизации
pop_size = np.arange(5,26)
iterations_num = []
x_values = []

# Для каждого из значений запускаем алгоритм оптимизации и сохраняем результирующие значения
for cur_pop_size in np.nditer(pop_size):
    result = differential_evolution(ackley, bounds, seed=42, popsize=cur_pop_size)
    iterations_num.append(result.nit)
    x_values.append([result.x[0],result.x[1]])

# Выводим результаты
plt.xlabel("Popsize")
plt.ylabel("Iterations")
plt.plot(pop_size, iterations_num, marker='o')
print('Результаты оптимизации:')
print(x_values)
```
