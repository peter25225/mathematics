```python
import matplotlib.pyplot as plt
import math
from sympy import Symbol, Derivative, sin, cos
# import sympy as sym
```

###### Уровень 0

Посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
x = Symbol('x')

# Задаем исходную функцию
function = x**5 + 4 * sin(2*x) + cos(3*x + 3)

# Получаем первую производную
deriv1 = Derivative(function).doit()
print('Первая производная:')
deriv1
```

```python
# Считаем значение в точке 1
print(f'Значение первой производной в точке 1: {deriv1.evalf(subs={x:1})}')
```

```python
# Получаем вторую производную
print('Вторая производная:')
deriv2 = Derivative(deriv1).doit()
deriv2
```

```python
# Считаем значение в точке 1
print(f'Значение второй производной в точке 1: {deriv2.evalf(subs={x:1})}')
```

###### Уровень 1
Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
# Задаем исходную функцию
function = (sin(2*x+1))**5

# Получаем первую производную
deriv1 = Derivative(function).doit()
print('Первая производная:')
deriv1
```

```python
# Получаем вторую производную
print('Вторая производная:')
deriv2 = Derivative(deriv1).doit()
deriv2
```

```python
# Строим график первой производной
plt.figure()
xi = -5

while xi <= 5:
    plt.plot(xi, deriv1.evalf(subs={x:xi}), color='green', marker='.')
    xi += 0.005
plt.grid()
```

```python
# Строим график второй производной
plt.figure()
xi = -5

while xi <= 5:
    plt.plot(xi, deriv2.evalf(subs={x:xi}), color='red', marker='.')
    xi += 0.005
plt.grid()
```
